```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, mode, skew, kurtosis

# Parameter distribusi normal
mu = 5          # Mean
sigma = 1     # Standar Deviasi (asumsi)
x_value = 3.5   # Nilai yang ingin dicari probabilitasnya
n = 1000        # Jumlah sampel (n > 100)
# Menghitung skor Z
z = (x_value - mu) / sigma
print(f"Z score untuk x = {x_value} adalah {z}")

# Menghitung probabilitas P(X < 3.5)
prob = norm.cdf(x_value, loc=mu, scale=sigma)
print(f"Probabilitas P(X < {x_value}) adalah {prob:.4f}")

# Membuat data acak dari distribusi normal
data = np.random.normal(loc=mu, scale=sigma, size=n)
```

```python
# Menghitung statistik deskriptif
mean = np.mean(data)
modus = mode(data, keepdims=True).mode[0]
median = np.median(data)
percentile_20 = np.percentile(data, 20)
percentile_80 = np.percentile(data, 80)
variance = np.var(data)
data_skewness = skew(data)
data_kurtosis = kurtosis(data)

# Menampilkan hasil
print(f"\nStatistik Deskriptif:")
print(f"Mean: {mean:.4f}")
print(f"Modus: {modus:.4f}")
print(f"Median: {median:.4f}")
print(f"Persentil ke-20: {percentile_20:.4f}")
print(f"Persentil ke-80: {percentile_80:.4f}")
print(f"Varians: {variance:.4f}")
print(f"Skewness: {data_skewness:.4f}")
print(f"Kurtosis: {data_kurtosis:.4f}")
```

```python
# Visualisasi Distribusi Normal
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = norm.pdf(x, loc=mu, scale=sigma)

plt.plot(x, pdf, label='Distribusi Normal')
plt.fill_between(x, pdf, where=(x < x_value), color='skyblue', alpha=0.5, label=f'P(X < {x_value}) = {prob:.4f}')
plt.title('Distribusi Normal: Waktu Maintenance Server')
plt.xlabel('Waktu (jam)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
# Plot CDF
plt.subplot(1, 2, 2)
x = np.linspace(min(data), max(data), 1000)
cdf = norm.cdf(x, loc=mu, scale=sigma)
plt.plot(x, cdf, color='green', label='CDF')
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
