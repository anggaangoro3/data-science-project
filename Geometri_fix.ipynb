```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, mode
import math

# Parameter yang diberikan
total_respondents = 120
old_laptops = 1
p = old_laptops / total_respondents  # Probabilitas seorang responden memiliki laptop jadul

k = 15  # Probabilitas bahwa laptop ke-15 adalah laptop jadul pertama

# a. Menghitung probabilitas menggunakan distribusi geometrik dengan scipy.stats
prob_scipy = stats.geom.pmf(k, p)


# Menampilkan hasil
print("probabil ke 10 Menggunakan scipy.stats:")
print(f"Probabilitas: {prob_scipy:.6f}")
print(f"Dalam Persen: {prob_scipy*100:.4f}%\n")
```

```python
# --- Membangkitkan Data Secara Acak dan Analisis Statistik Deskriptif ---

# c. Membangkitkan data
n = 500 # Jumlah data
data = stats.geom.rvs(p, size=n)

# d. Menghitung Statistik Deskriptif
mean_val = np.mean(data)
median_val = np.median(data)
modus = stats.mode(data, keepdims=True)[0][0]
# Menghitung persentil ke-25 dan ke-75
percentile_25 = np.percentile(data, 25)
percentile_75 = np.percentile(data, 75)
variance_val = np.var(data)
std_dev = np.sqrt(variance_val)
skewness = skew(data)
kurt_val = kurtosis(data)

# Menampilkan hasil
print(f"Mean (Rata-rata): {mean_val:.2f}")
print(f"Median: {median_val}")
print(f"Mode: {modus}")
# Menampilkan hasil
print(f"Persentil ke-25: {percentile_25: .2f}")
print(f"Persentil ke-75: {percentile_75: .2f}")
print(f"Varians: {variance_val:.2f}")
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurt_val:.2f}\n")
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.histplot(data, bins=20, kde=False, stat="probability", color='skyblue')  # Adjust bins as needed
plt.title('Probabilitas Distribusi Geometrik')
plt.xlabel('Jumlah percobaan hingga sukses pertama (k)')
plt.ylabel('Probabilitas')
plt.grid(True)

plt.show()
```
